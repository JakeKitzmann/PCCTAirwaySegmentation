```python
import urllib.request, zipfile, os, glob

url = "https://zenodo.org/records/10069289/files/AeroPath.zip?download=1"
if not os.path.exists("AeroPath.zip"):
    urllib.request.urlretrieve(url, "AeroPath.zip")

with zipfile.ZipFile("AeroPath.zip") as z:
    z.extractall("AeroPath")

for p in sorted(glob.glob("AeroPath/**/*.nii.gz", recursive=True))[:10]:
    print(p)
```
